**Problem 4**. Read in the Houses data file using the following code.  The variables include price (thousands of dollars), size (square feet), tax bill (dollars), number of bathrooms, number of bedrooms, and whether the house is new (1 = yes, 0 = no) for 100 home sales in Gainesville, Florida. Let
y = selling price and x = size of home.

```python
houses = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Houses.dat")
display(houses)
```

Part a. Construct a scatterplot between square feet and price of a home. 

```python
x = houses['size']
y = houses['price']

plt.scatter(x,y)
plt.show()
```

Part b. Find the correlation between the two variables.

```python
houses['size'].corr(houses['price'])
```

Part c. Find and plot the regression line on top of the scatterplot. 

```python
from sklearn.linear_model import LinearRegression
```

```python
x = houses[['size']]
y = houses['price']
model = LinearRegression()
model.fit(x,y)

slope = model.coef_[0]
intercept = model.intercept_

plt.scatter(x,y, alpha=.6)
plt.plot(x, model.predict(x), color='red', label='Regression Line')
```

Part d. Create a side by side boxplot of price comparing homes that are new and homes that are not new. 

```python
houses.boxplot(column='price', by='new', vert=False, grid=False)
```

Part e. What is the relationship between size and price.  Justify your answer using the plots created. Do the same for the relationship between price and whether or not a house is new, again using the plot to justify your answer.

The relationship between size and price is a strong positive correlation. As size goes up, price tends to go up as well. This can be seen in the scatter plot.

Additionally, the newer a home is, the more likely it is to be higher priced. This can be seen in the box plot that shows price distribution with new and old houses. New houses are spread through a much higher range of price.

**Problem 1.22**. Using the Happy data file read in by the following code. Construct the contingency table relating marital status and happiness. 

```python
happy = pd.read_csv("https://stat4ds.rwth-aachen.de/data/Happy.dat")
happy.head(20)
```

```python
happy['happiness'] = [int(h.split()[1]) for h in happy['subject happiness marital gender']]
happy['marital_status'] = [int(h.split()[2]) for h in happy['subject happiness marital gender']]
happy['gender'] = [h.split()[3] for h in happy['subject happiness marital gender']]
display(happy.head(20))
```

```python
# continugency table between marital status and happiness
pd.crosstab(happy['marital_status'], happy['happiness'])
```
